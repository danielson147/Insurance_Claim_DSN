```python
# Model Building
from sklearn import metrics
 
from sklearn.model_selection import train_test_split
```

```python
# Now it is time to split data into features and target
# Our label is the total_claim_amount
X = df_train.drop(["total_claim_amount"], axis="columns")
y = df_train["total_claim_amount"]
```

```python
# Let us then views our features
X.head()
```

```python
 # Let us then views our label
y.head()
```

 Our label is continuous, therefore the task of predicting claims for the insurance firm is a regression task

 We are going to do one-hot encoding for all the categorical features. This is because some algorithms cannot work with categorical data directly. They required all input variables and output variables to be numeric. We will create one-hot encoding for insured_sex, policy_state, property_damage, collision_type

```python
#We need to drop the Customer_ID features
X = df_train.drop(["Customer_ID"], axis="columns")
```

```python
X.head(5)
```

```python
# We will now use pd.get_dummies() to create one-hot encoding
pd.get_dummies(X, prefix=["insured_sex", "police_report_available"], columns=["insured_sex", "police_report_available"])
```

```python
#We will then save the result of one-hot encoding into X
X = pd.get_dummies(X, prefix=["insured_sex", "police_report_available"], columns=["insured_sex", "police_report_available"])
X.head()
```

```python
# We will split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1234)
```

```python
# Let us view the shape of our data
X.shape
```

```python
# We drop every other column except the ones we need for training the model 
X_train = X_train[['insured_sex_FEMALE', 'insured_sex_MALE', 'police_report_available_NO', 'police_report_available_YES']]
```

```python
X_train.head(5)
```

```python
#We need to drop the Customer_ID feature for the test set
df_test = df_test.drop(["Customer_ID"], axis="columns")
```

```python
#We will then save the result of one-hot encoding into df_test
df_test_1 = pd.get_dummies(df_test, prefix=["insured_sex", "police_report_available"], columns=["insured_sex", "police_report_available"])
df_test_1.head()
```

```python
# We drop every other column except the ones we need for training the model 
X_test = df_test_1[['insured_sex_FEMALE', 'insured_sex_MALE', 'police_report_available_NO', 'police_report_available_YES']]
```

```python
X_test.head(5)
```

```python
# We need to train a linear regression model with our train data. 
# We will first import the Linear Regression from sklearn model
# Fitting linear regression to our training set
from sklearn.linear_model import LinearRegression
```

We now create an object of class LinearRegression to train the model on 

```python
# We need to create an object of LinearRegression to train the model on
linearmodel = LinearRegression()
 
linearmodel.fit(X_train, y_train)
```

```python
linearmodel.predict(X_test)
```

```python
# Let us now save our prediction result into linearmodel_prediction. This is what the model predicted for us.
linearmodel_prediction = linearmodel.predict(X_test)
```

```python
#We will now save our prediction to csv
# np.savetxt('insurance_prediction.csv', linearmodel_prediction, delimiter = ",")
```

```python
claim_prediction = pd.DataFrame(linearmodel_prediction)
claim_prediction.to_csv("/content/drive/MyDrive/DSN_Hackathon/claims.csv")
```
